```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
plt.ion()
```

```python
df= pd.read_excel("iris_rev.xlsx")
```

```python
sns.relplot("slength", "plength", hue="pwidth", palette="ch:r=.4",style="species", data=df)
```

```python
sns.palplot(sns.cubehelix_palette(rot=-.4))
```

```python
sns.palplot(sns.cubehelix_palette(start=2.8, rot=.4))
```

```python
a=sns.catplot("species", "plength",kind="swarm", hue="species", palette="ch:r=5.5", data=df)
```

```python
a.set_axis_labels("","Petal Length (mm)")
a.savefig("swarm.png")
```

```python
sns.stripplot("species", "slength", data=df, jitter=1)
```

```python
sns.stripplot("slength","species",data=df,jitter=1, linewidth=1)
```

```python
plot2=sns.stripplot("species", "slength", hue="region", data=df, palette="ch:r=.4", dodge=True)
plt.xlabel("")
plt.ylabel("Sepal Length (mm)")
plt.show(plot2)
```

```python
fig=plot2.get_figure()
```

```python
fig.savefig("stripplot.png")
```
